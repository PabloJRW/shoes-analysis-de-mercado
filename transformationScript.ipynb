```python
import os
import pandas as pd
import datetime as dt
```

```python
FILE_PATH = os.path.join('dataset','shoes.csv')
```

```python
df = pd.read_csv(FILE_PATH, index_col=False)
df
```

```python
# Cambio de tipado a fecha
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%m/%d/%y")
# Cambio de tipado a fecha
df['Release Date'] = pd.to_datetime(df['Release Date'], format="%m/%d/%y")
# Cambio de tipado a flotante
df['Sale Price'] = df['Sale Price'].str.replace('$','').str.replace(',','').astype(dtype='float')
# Cambio de tipado a flotante
df['Retail Price'] = df['Retail Price'].str.replace('$','').str.replace(',','').astype(dtype='float')

#
df['Sneaker Name'] = df['Sneaker Name'].str.lower().str.replace('-',' ').replace('\s+', ' ', regex=True)
```

```python
def extraer_marca(df, lista_de_marcas):
    # Itera sobre cada marca en la lista de marcas
    for marca in lista_de_marcas:
        # Comprueba si la marca esta presente
        if marca in df:
            return marca
    # Si ninguna marca está presente, retorna "nike"
    return "nike"


lista_de_marcas = ['nike', 'adidas']
# Aplica la función "extraer_marca"
df['Marca'] = df['Sneaker Name'].apply(extraer_marca, lista_de_marcas=lista_de_marcas)
```

```python
def extraer_modelo(df, lista_de_modelos):
    
    for modelo in lista_de_modelos:
        df = df.apply(lambda x: modelo if modelo in x else x)

    return df


# Lista de los modelos a buscar
lista_de_modelos = ["yeezy boost 350 v2", "air force 1", "zoom fly", "zoom fly mercurial", "air max 97", 
                    "air max 90", "blazer mid", "air presto", "yeezy boost 350 low", "air vapormax", 
                    "air jordan 1 retro", "air jordan retro 1", "react hyperdunk"]

# Aplica la función "extraer_modelo"
df['Modelo'] = extraer_modelo(df['Sneaker Name'], lista_de_modelos)
```

```python
mis_variables = ['Release Date', 'Order Date', 'Marca', 'Modelo', 'Shoe Size', 'Retail Price', 'Sale Price', 'Buyer Region']
df = df[mis_variables].copy()

nombres_de_columnas = {'Release Date': 'Fecha de Lanzamiento', 'Order Date': 'Fecha de Compra', 'Shoe Size': 'Talla',
                       'Retail Price': 'Precio Sugerido', 'Sale Price': 'Precio de Venta', 'Buyer Region': 'Región'}

df = df.rename(columns=nombres_de_columnas, inplace=True)
```
